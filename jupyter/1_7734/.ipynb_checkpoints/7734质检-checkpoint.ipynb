# 20年1月8日

```python
f = mytools.othern.file_name_paths('G:/1-规划/7734/7734质检/20200108数据质检7734/数据/',find='.csv')
```

```python
data = pd.DataFrame()
for name in f:
    data_t = pd.read_csv(open(name,encoding='utf-8'),sep='\t')
    data = data.append(data_t)
```

```python
data.groupby('day').count()
```

```python
data_res = data.groupby(by = 'eutrancell_cgi').sum()
data_res.reset_index(inplace = True)
data_res.reset_index(inplace = True)
rew = data_res.merge(data,how = 'left' , on = 'eutrancell_cgi')
res = rew.drop_duplicates('index')
```

```python
res = res[['eutrancell_cgi', 'rsrp_sample_count_x','rsrp_weak_x','region_name']]
res['fgl'] = 1-res['rsrp_weak_x']/res['rsrp_sample_count_x']
```
