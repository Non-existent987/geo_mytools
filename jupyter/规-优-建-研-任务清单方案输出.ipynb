```python
def changjing_distance(df,changjing,zhishi):
        F900 = {0:'300',1:'430',2:'430',3:'430',4:'1200',5:'2200'}
        TDD = {0:'250',1:'400',2:'400',3:'350',4:'800',5:'1000'}
    
        df['juli']= [int(F900[cj]*2) if pd == 'FDD900' else int(TDD[cj])*2
                 for cj , pd in zip(df[changjing],df[zhishi])]
        return df
```

```python
cj_900_hxcq_2_x = 111
cj_900_ybcq_2_x = 222
cj_900_cj_2_x = 333
cj_900_xc_2_x = 444
cj_900_xz_2_x = 555
cj_900_nc_2_x = 666

cj_1800_hxcq_2_x = 1111
cj_1800_ybcq_2_x = 665
cj_1800_cj_2_x = 345
cj_1800_xc_2_x = 2342
cj_1800_xz_2_x = 435435
cj_1800_nc_2_x = 234
```

```python
def juli_2(df, changjing, pinduan,beishu=1):
    F900 = {0: cj_900_hxcq_2_x, 1: cj_900_ybcq_2_x, 2: cj_900_cj_2_x, 3: cj_900_xc_2_x, 4: cj_900_xz_2_x, 5: cj_900_nc_2_x}
    TDD = {0: cj_1800_hxcq_2_x, 1: cj_1800_ybcq_2_x, 2: cj_1800_cj_2_x, 3: cj_1800_xc_2_x, 4: cj_1800_xz_2_x, 5: cj_1800_nc_2_x}
    df['juli'] = [F900[cj]*beishu if pd == 'FDD900' else TDD[cj]*beishu 
                    for cj, pd in zip(df[changjing], df[pinduan])]
    df['juli'] = pd.to_numeric(df['juli'])
    return df
```
