```python
import mytools
import geopandas as gpd
import pandas as pd
```

```python
hb_tu = mytools.gisn.maps('湖北')
```

```python
changjing = gpd.read_file('g:/1-规划/图层/全省区域规划图层20180510/全省区域规划图层20180510/全省区域规划图层20180510.TAB')
```

```python
changjing_use = changjing[changjing['区域类型'] == '主城区']
```

```python
f = mytools.othern.file_name_paths(path='g:/1-规划/图层/湖北建筑/',find='.tab',case_sensitive=False)
```

```python
mianji = list()
gaodu = list()
dishi = list()
for name in f:
    louyu_t = gpd.read_file(name)
    louyu_t = louyu_t.to_crs({'init': 'epsg:4326'})
    res = gpd.sjoin(louyu_t,changjing_use)
    res_area = mytools.gisn.add_area(res,column='面积1')
    mianji.append(res_area['面积1'].sum())
    dishi.append(name.split('湖北建筑\\湖北建筑\\')[1].split('.')[0])
    gaodu.append(res_area['建筑物高度'].mean())
    print(name,'ok')
    print(dishi,mianji,gaodu)
```

```python
data = pd.DataFrame()
data['city']=dishi
data['高度']=gaodu
data['面积']=mianji
```

```python
data['city'] = data['city'].replace({'仙桃':'江汉','神农架':'十堰'})
```

```python
data_res = data.merge(changjing_use,how='left',left_on='city',right_on='地市名称')
```

```python
data_res1 = data_res[['city', '高度', '面积', '覆盖面积']]
```

```python
data_res1.columns = ['city', '高度', '楼宇面积', '总面积']
```
