```python
import Ipynb_importer
import kml-kmlpolygon
import pandas as pd
rr = pd.read_csv("rrr.csv",encoding='gbk')
rr2 = rr[:]
rr1= rr2.copy()
jingweidu=[]
for i in range(len(rr1.index)):
    row5 = rr1.iloc[i,:]
    row_z5 = row5[['jingdu','weidu','fangweijiao','juli']] # +++++++++++++++++++++++做多边形使用列
    jingdu,weidu,fangweijiao,juli=row_z5[['jingdu','weidu','fangweijiao','juli']]
    x = kmlpolygon._point_lon_lat(lon=jingdu,lat=weidu,azimuth=fangweijiao,radius=juli)
    jingweidu.append(x)
rr1['jingweidu'] = jingweidu
csv1=rr1[['id','jingweidu']]
csv1.to_excel('jingweidu.xls',index=False) #将数据重新写入excel
csv1
```
